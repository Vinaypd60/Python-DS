# What is Data-Types?
## The main purpose of the data-types is that to allocate sufficient amount of memory space for inputs of the program.
## The values are called Data-types.(var name = values)

## Integer\whole\integral values are nothing but values without decimal points.

# Complex:
## The purpose of complex DT is that to store imaginary data and complex number.
## The general format of complex number a + bj (or) a - bj
### Here a is called real part and b is called img part and j is sqrt(-1)

```python
a = 3 + 4j
print(a,type(a))
```

```python
a = 3 - 4j
print(a,type(a))
```

```python
b = 4.5 - 5.6j
print(b,type(b))
```

```python
a = 2 + 4.5j
print(a)
print(a.real)
print(a.imag)
```

```python
b = 4.4 + 6.5j
print(b)
print(b.real)
print(b.imag)
```

```python
a = True + 4.5j
print(a,type(a))
```

```python
a = 0b1010 - 5j
print(a,type(a))
```

```python
c = 5 +6j
c1 = 6 +8j
print(c + c1)
```
