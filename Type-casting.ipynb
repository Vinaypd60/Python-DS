# The process of converting one possible type of value into another possible type of values is called "Type casting".
# In python, We have 5-fundamental type casting techniques. they are
## 1) Int()
## 2) Float()
## 3) bool()
## 4) complex()
## 5) str()

# Integer()

```python
int(112.89)
```

```python
int(True)
```

```python
int(False)
```

```python
int(True, False) # type error
```

```python
int(10+2j) # type error
```

```python
int('123')
```
